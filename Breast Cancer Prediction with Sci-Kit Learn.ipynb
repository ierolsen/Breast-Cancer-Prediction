# Breast Cancer Dataset

* M = malignant = 1 
* B = benign = 0

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data.drop(['id','Unnamed: 32'],axis=1, inplace=True)

data.head()
```

```python
M = data[data.diagnosis == 'M']
B = data[data.diagnosis == 'B']
```

# Scatter Plot

```python
plt.scatter(M.radius_mean, M.texture_mean, color = 'red', label = 'Malignant',alpha=0.2)
plt.scatter(B.radius_mean, B.texture_mean, color = 'green', label = 'Benign',alpha=0.2)
plt.xlabel('radius_mean')
plt.ylabel('texture_mean')
plt.legend()
plt.show()
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1,'B':0})
```

```python
data.head()
```

```python
y = data.diagnosis.values
X = data.drop(['diagnosis'],axis=1)
```

```python
y
```

# Normalization 

```python
X = (X - np.min(X))/(np.max(X)-np.min(X)).values
```

```python
X.head()
```

# Creating Train and Test Set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

# Transpose

```python
X_train = X_train.T
X_test = X_test.T

y_train = y_train.T
y_test = y_test.T
```

# Prediction with Logistic Regression

```python
from sklearn import linear_model
logreg = linear_model.LogisticRegression(random_state = 42,max_iter= 150)
print("test accuracy: {} ".format(logreg.fit(X_train.T, y_train.T).score(X_test.T, y_test.T)))
print("train accuracy: {} ".format(logreg.fit(X_train.T, y_train.T).score(X_train.T, y_train.T)))
```

# Prediction with Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

ran_for = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)
print("test accuracy: {} ".format(ran_for.fit(X_train.T, y_train.T).score(X_test.T, y_test.T)))
print("train accuracy: {} ".format(ran_for.fit(X_train.T, y_train.T).score(X_train.T, y_train.T)))
```

# Prediction with KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train.T, y_train.T)
prediction = knn.predict(X_test.T)
prediction
```

```python
print('score: ',format(knn.score(X_test.T, y_test.T)))
```

# Find k Values

```python
score = []

for i in range(1,15):
    knn2 = KNeighborsClassifier(n_neighbors=i)
    knn2.fit(X_train.T, y_train.T)
    score.append(knn2.score(X_test.T, y_test.T))
    
plt.plot(range(1,15), score)
plt.xlabel('k values')
plt.ylabel('sccuracy')
plt.show()
```

# Prediction with SVM

```python
from sklearn.svm import SVC

svm = SVC(random_state=42)
svm.fit(X_train.T, y_train.T)
print('score: ',svm.score(X_test.T, y_test.T))
```

#  Prediction with Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_train.T, y_train.T)
print('score: ',nb.score(X_test.T, y_test.T))
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier()
dt.fit(X_train.T,y_train.T)
print('score: ',dt.score(X_test.T, y_test.T))
```
